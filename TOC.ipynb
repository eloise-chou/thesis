# 介紹

近年來政府積極推動大眾運輸系統的發展，以減少交通擁擠和環境污染。然而，儘管大眾交通運輸系統已發展完善，但自有車的便利性仍然促使汽機車數量持續增加。大量的汽機車流量加上台灣許多道路都沒有足夠的行人通道，導致行人在道路上行走時面臨著很大的風險。這種情況使得台灣被稱為「行人地獄」。

此研究中我將納入交通流量、道路是否有專屬左轉道、是否有機車待轉區等等道路資訊，去分析交通事故發生次數與交通流量間因果關係

# 結論

透過 Double-Lasso 回歸後得到以下結果，當路口涵蓋範圍較小時(250 m)，交通流量每增加 1% 會導致交通事故次數增加 0.33%，而當路口涵蓋範圍擴大後(500m)，影響上升至交通流量每增加 1% 會導致交通事故次數增加 0.41% 。這可能是因為在路口涵蓋範圍較大的情況下，會受到不屬於該路口的交通事故特性的影響，導致Double-Lasso在挑選控制變數時未能將重要的中間變數「機車待轉區」納入回歸模型，進而使交通流量對交通事故次數的影響呈現高估的情況。

理想的交通情況應該是透過道路工程去將流量增加對事故發生的影響降低或降為 0，但實際情況卻是流量的增加經過道路工程後，仍然對事故發生有直接的影響。因此相關的部門可以往將道路工程做的更完善為目標去努力。
