# 整理路口資料

將分時的車流量整理為「上午_機車」、「上午_汽車」、「上午_卡車」、「下午_機車」、「下午_汽車」、「下午_卡車」

程式碼請參考 [model/get_traffic_intersection](./model/get_traffic_intersection.py)

```python
traffics:List[Traffic] = []
error_intersections = []
for intersection in tqdm(intersection_list):
    try:
        tr = get_traffic_table(f'DATA/traffic/{intersection}.xlsx')
        tr.extract_traffic_data()
        traffics.append(tr)
    except:
        error_intersections.append(intersection)

traffic_df = pd.DataFrame([tr.to_dict for tr in traffics])
traffic_df.to_csv('車流量整理.csv', index = False, encoding='utf-8')
```
